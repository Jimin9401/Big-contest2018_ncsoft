### Customize  ,  trans primitive
#### 1.   

```python
def string_count(column, string=None):
    assert string is not None, "string to count needs to be defined"
    counts = [element.lower().count(string) for element in column]
    return counts
```

#### 2.   column           

```python
def string_count_generate_name(self):
    return u"STRING_COUNT(%s, %s)" % (self.base_features[0].get_name(),'"'+str(self.kwargs['string']+'"'))
```

#### 3. function  trans primitive  

```python
StringCount = ft.primitives.make_trans_primitive(function=string_count,
                                   input_types=[Text],
                                   return_type=Numeric,
                                   cls_attributes={"generate_name": string_count_generate_name})
```
