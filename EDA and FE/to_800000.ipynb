```python
import pandas as pd
```

```python
#timeseries
activity=pd.read_csv('lite_data/train_activity_lite.csv')
label=pd.read_csv('lite_data/train_label_lite.csv')
```

```python
def fill_na_data(data,user_num=100000,replace_text='tr'):
    activity_numeric_id=data.copy()
    activity_numeric_id['new_id']=activity_numeric_id['new_id'].str.replace(replace_text,'').tolist()
    activity_numeric_id['new_id']=pd.to_numeric(activity_numeric_id['new_id'])
    minvalues=[]
    for i in activity_numeric_id.columns[2:]:
        smallest_values=activity_numeric_id[i].value_counts().index.sort_values()
        minvalues.append(smallest_values[:][0]-abs(smallest_values[:][0]-smallest_values[:][1]))
    minvalues_final=pd.DataFrame([minvalues,minvalues],columns=activity_numeric_id.columns[2:]).iloc[0,:]
    #number of users:
    aa=[[a//8,(a%8)+1] for a in range(user_num*8)]
    activity_80=pd.DataFrame(aa)
    activity_80.columns=['new_id','wk']
    activity_80=pd.merge(activity_80,activity_numeric_id,how='outer')
    for i in activity_80.columns[2:]:
        activity_80[i]=activity_80[i].fillna(value=minvalues_final[i])
    return activity_80
    
```

```python
adad=fill_na_data(activity)
```
