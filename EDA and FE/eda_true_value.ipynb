# 만든 셋으로 decision tree score 보기

```python
X_train=datas_int.groupby('new_id').mean()
X_train.head()
```

```python
len
```

```python
label_original=pd.read_csv('lite_data/train_label_lite.csv')
label_original.head()
```

```python
#score를 저장할 공간
score=[]
score_col_name=X_train.columns

for i in range(len(X_train.columns)):
    print(X_train.columns[i])
    clf = clf.fit(X_train.iloc[:,i:i+1], label_original.label)
    score.append(clf.score(X_train.iloc[:,i:i+1], label_original.label))
```

```python
score
```

```python
#score data를 dataframe으로 만들어 순위를 확인
score_copy=score.copy()
score_dt=pd.DataFrame(np.array([score_col_name,score_copy]).transpose())
```

```python
score_dt[1].sort_values(ascending=False).head(50)
```

```python
score_dt.to_csv('interaction_tree_score.csv',index=False)
```
