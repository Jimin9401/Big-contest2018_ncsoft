```python
import numpy as np
import tensorflow as tf

def get_input_fn(x_data, y_data, batch_size, capacity=50000, min_after_dequeue=3000):
    def _input_fn():
        x_batch, y_batch = tf.train.shuffle_batch(
            tensors=[x_data, y_data.astype(np.int32)],
            batch_size=batch_size,
            capacity=capacity,
            min_after_dequeue=min_after_dequeue,
            enqueue_many=True,
            num_threads=4)
        features_map = {'inputs': x_batch}
        return features_map, y_batch
    return _input_fn

X_train=train_np.astype(np.float32)
y_train=label_np.astype(np.float32)
X_val=train_np.astype(np.float32)
y_val=label_np.astype(np.float32)

train_input_fn = get_input_fn(X_train,y_train, batch_size=10000)
eval_input_fn = get_input_fn(X_val,y_val, batch_size=5000)

# Specify the feature(s) to be used by the estimator. This is identical to the
# code used for the LinearClassifier.
x_data=tf.contrib.layers.real_valued_column('inputs',dimension=feature_num,normalizer=(lambda x:tf.layers.batch_normalization(x,training=True)))
optimizer = tf.train.FtrlOptimizer(learning_rate = 5., l2_regularization_strength=0.001)

#gpu setting
config = tf.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 2.0
config._save_summary_steps: 100



kernel_mapper = tf.contrib.kernel_methods.RandomFourierFeatureMapper(
  input_dim=feature_num, output_dim=to_dim, stddev=5.0, name='rffm')
kernel_mappers = {x_data: [kernel_mapper]}
estimator = tf.contrib.kernel_methods.KernelLinearClassifier(n_classes=4, optimizer=optimizer, kernel_mappers=kernel_mappers,
                                                             config=tf.estimator.RunConfig(session_config=config))
```

```python
# Train.
start = time.time()
estimator.fit(input_fn=train_input_fn, steps=100000)
end = time.time()
print('Elapsed time: {} seconds'.format(end - start))

# Evaluate and report metrics.
eval_metrics = estimator.evaluate(input_fn=eval_input_fn, steps=1)
print(eval_metrics)
```
